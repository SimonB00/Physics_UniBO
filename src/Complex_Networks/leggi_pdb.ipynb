```python
import scipy as sp
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
A = sp.io.loadmat('coord1PHP.mat')
A = A['protdist'][:,0]
Nnode = len(A)

sns.histplot(A, kde=True, bins=50)
```

```python
D = sp.spatial.distance.squareform(A)
sns.heatmap(1./D)
```

```python
print(D.shape)
Net6 = (D < 6) * D
print(Net6.shape)
Net10 = (D > 10) * (D < 11) * D
print(Net10.shape)
N1 = (D < 11) * D
sns.heatmap(Net6 + 2*Net10)
```

```python
Kdiag = []

for index in range(20):
    Kdiag.append(np.sum(np.diag(A, index))/(Nnode - index))

plt.plot(np.log(Kdiag), '.')
```
