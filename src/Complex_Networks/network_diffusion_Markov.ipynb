# Markov process

```python
import scipy as sp

data = sp.io.loadmat('Nets_Diffusion_10.mat')
A = data['Block10_7_3']
Nnode = len(A)
```

```python
import numpy as np

K = np.sum(A, axis=0)
T = np.matmul(A, np.linalg.inv(np.diag(K)))
```

Now set the initial conditions...

```python
x = np.random.rand(Nnode)
x0 = float(np.max(x))

from matplotlib import pyplot as plt

plt.bar(range(Nnode), x)
```

```python
Nit = 30

for ind in range(Nit):
    x = T.dot(x)
    
plt.bar(range(Nnode), x)
```
